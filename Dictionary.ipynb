```python
"""
Author: Agasthya Vidyanath Rao Peggerla
"""

fname = open('alphabet.txt')
fout = open('output.txt', "w")

# A dictionary to store the letter and its predecessors
# key = letter, value = it's predecessors(not in order)

alphabet_map = dict()

#Process the first line

line1 = fname.readline()

#Remove spaces

line1 = line1.strip()

for i in range(len(line1)):
    #Initialise the first letter of the alphabet with no predecessors
    if i == 0:
        alphabet_map[line1[0]] = set()
    else:
    # Insert the remaining letters from the first line into dictionary with the first letter as predecessor
        if line1[i] != line1[0]:
            alphabet_map[line1[i]] = set()
            alphabet_map[line1[i]].add(line1[0])

line2 = fname.readline()

#Read the file till end of file
while line2:

    line2 = line2.strip()
    
    for i in range(len(line2)):
        if (i<len(line1)):
            
            #Check for any mismatch in the position of characters from the previous word
            #and insert into the dictionary accordingly
            
            if (line1[i] != line2[i]):
                # If there is an entry for the current letter in the dictionary update it's value
                if line2[i] in alphabet_map:
                    alphabet_map[line2[i]].add(line1[i])
                    break
                else:
                # Else enter the current letter into the dictionary and update it's value
                    alphabet_map[line2[i]] = set()
                    alphabet_map[line2[i]].add(line1[i])
                    break
                
    line1 = line2                                 
    line2 = fname.readline()

fname.close()

# A list to hold the letters in order

result_list = [''] * len(alphabet_map)

map_set = set()

# For each entry in the dictionary find all the predecessors by iterating recursively
for key,val in alphabet_map.items():
    map_set = val
    new_set = set()
    new_set = map_set
    
    for x in new_set:
        # Find out all the predecessors of the current letter by iterating over it's predecessors
        map_set = map_set.union(alphabet_map.get(x))
        
    # Update the predecessor value for the current letter    
    alphabet_map[key] = map_set
    
    # Insert the letter into the list at an index based on the number of predecessors
    result_list[len(map_set)] = key

# Writing the output to a file
for i in range(len(result_list)):
    letter_num = i+1
    line_write = "Letter " + str(letter_num) + " => '" +str(result_list[i])+"'\n"
    fout.write(line_write)
fout.close()

```
